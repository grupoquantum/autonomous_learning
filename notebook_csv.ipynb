<h2>Autonomous Learning (Aprendizado Autônomo/Aprendizado NÃO Supervisionado)</h2>

<h3>Exploração dos Dados</h3>

```python
from Neuraline.Utilities.data import DataTable, DataExploration
from Neuraline.Utilities.chart import Chart
data_table, data_exploration, chart = DataTable(), DataExploration(), Chart()
url_path = './bitcoin01.csv'
data_table.visualization_csv_inline(url_path=url_path, limit=10)
```

<h3>Separação em Treinamento e Teste</h3>

```python
y_training, _, y_test, _ = data_exploration.dataSeparationCSV(url_path=url_path, list_inputs=['Fechamento'], training_percentage=.5, test_percentage=.5)
print(f'y_training: {len(y_training)} - y_test: {len(y_test)}')
```

<h3>Visualização dos Dados de Treinamento</h3>

```python
chart.plotLINE(y=y_training, y_label='Fechamento', title='Cotação do Bitcoin - Dados de Treinamento')
```

<h3>Visualização dos Dados de Teste</h3>

```python
chart.plotLINE(y=y_test, y_label='Fechamento', title='Cotação do Bitcoin - Dados de Teste')
```

<h3>VEC/VECM</h3>

```python
from Neuraline.ArtificialIntelligence.MachineLearning.AutonomousLearning.vector_error_correction import VectorErrorCorrection
vector_error_correction = VectorErrorCorrection()

result = vector_error_correction.fitPredict(inputs=y_training, predictions=116)
chart.plotMATRIX(matrix1=y_test, matrix2=result)
```

```python
vector_error_correction.fitCSV(url_path=url_path)
```

```python
vector_error_correction.predictCSV(url_path='previsao_bitcoin.csv', titles=['Abertura', 'Fechamento', 'Mínima', 'Máxima'], predictions=20, precision=1)
data_table.visualization_csv_inline(url_path='previsao_bitcoin.csv', limit=10)
```

```python
vector_error_correction.fitPredictCSV(url_path='./bitcoin02.csv', predictions=10, precision=1)
data_table.visualization_csv_inline(url_path='./bitcoin02.csv', limit=10)
```
