<h2>Autonomous Learning (Aprendizado Autônomo/Aprendizado NÃO Supervisionado)</h2>

<h3>Exploração dos Dados</h3>

```python
from Neuraline.Utilities.data import DataTable, DataExploration
from Neuraline.Utilities.chart import Chart
data_table, data_exploration, chart = DataTable(), DataExploration(), Chart()
url_path = './PETR4 Dados Históricos.csv'
data_table.visualization_csv_inline(url_path=url_path, limit=10)
```

<h3>Separação em Treinamento e Teste</h3>

```python
y_training, _, y_test, _ = data_exploration.dataSeparationCSV(url_path=url_path, list_inputs=['Fechamento'], training_percentage=.8, test_percentage=.2)
print(f'y_training: {len(y_training)} - y_test: {len(y_test)}')
```

<h3>Visualização dos Dados de Treinamento</h3>

```python
chart.plotLINE(y=y_training, y_label='Fechamento', title='Cotação PETR4 - Dados de Treinamento')
```

<h3>Visualização dos Dados de Teste</h3>

```python
chart.plotLINE(y=y_test, y_label='Fechamento', title='Cotação PETR4 - Dados de Teste')
```

<h3>VEC/VECM</h3>

```python
from Neuraline.ArtificialIntelligence.MachineLearning.AutonomousLearning.vector_error_correction import VectorErrorCorrection
vector_error_correction = VectorErrorCorrection()

predictions = 19
result = vector_error_correction.fitPredict(inputs=y_training, predictions=predictions, preprocessor='standardization')
chart.plotMATRIX(matrix1=y_test[:predictions], matrix2=result)
```

```python
vector_error_correction.fitPredictCSV(url_path='./petr4.csv', predictions=19, precision=1)
data_table.visualization_csv_inline(url_path='./petr4.csv', limit=19)
```
