<h3>Redes Neurais de Hopfield</h3>
<p>Redes de Hopfield são redes neurais artificiais recorrentes baseadas na arquitetura de memorização humana com memória de recuperação, geralmente são utilizadas na recuperação de arquivos corrompidos e na remoção de ruídos.</p>

<h4>Importação das Classes e Instanciação dos Objetos</h4>

```python
from Neuraline.ArtificialIntelligence.MachineLearning.AutonomousLearning.hopfield_network import HopfieldNetwork
from Neuraline.Utilities.chart import Chart
hopfield_network, chart = HopfieldNetwork(), Chart()
```

<h4>Treinamento da Rede Neural de Hopfield</h4>

```python
hopfield_network.addFit(url_path='./images/training/', progress=True)
```

<h4>Plotagem para a Visualização da Imagem de Entrada da Predição</h4>

```python
url_input = './images/test/image04.jpg'
chart.plotIMAGE(url_input)
```

<h4>Predição e Visualização do Resultado</h4>

```python
result = hopfield_network.predict(url_input)
chart.plotIMAGE('./predicted.jpg')
```
